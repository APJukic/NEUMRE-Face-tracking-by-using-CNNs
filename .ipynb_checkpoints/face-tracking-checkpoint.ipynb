# Face tracking by using CNNs

```python
from PIL import Image
import torch
import torchvision
```

### DATASET


DATASET from WIDER FACE: A Face Detection Benchmark

link: https://shuoyang1213.me/WIDERFACE/

    download WIDER Face Training Images 



images used for training: 
    
    9--Press_Conference
    
image info in file:
    
    press-conference.txt   -- contains bounding box of face for each image (left, top, width, height)


```python
# info for bounding box for all images saved in images_info list
path = "press-conference.txt"

with open(path) as f:
    lines = f.readlines()

images_info = []


i = 0
while i < len(lines):
    tmp = lines[i+1]

    if int(tmp)==1: #one person per image
        image_info = []
        
        name = lines[i]    #image name
        bounds = lines[i+2].split(" ")
        x = int(bounds[0])  # x to left
        y = int(bounds[1])  # top
        width = int(bounds[2]) #width of face
        height = int(bounds[3]) #height of face
        
        image_info.extend([name.strip(), x, y, width, height])
        images_info.append(image_info)

        i = i + 3
    else:
        i = i + int(tmp) + 2
        pass
```

```python
#display of first image and bounding box around the face

from torchvision.io import read_image
from torchvision.utils import draw_bounding_boxes
print("FIRST IMAGE IN DATASET")
print("IMAGE INFO: NAME, x, y, width, height")
print(images_info[0])


path = images_info[0][0]
img = read_image(path)


box = [int(images_info[0][1]), int(images_info[0][2]) 
                                    , int(images_info[0][3]) + int(images_info[0][1])
                                    , int(images_info[0][4]) + int(images_info[0][2])]

box = torch.tensor(box)
box = box.unsqueeze(0)
  

img = draw_bounding_boxes(img, box, width=6,
                          colors="red")
  

img = torchvision.transforms.ToPILImage()(img)
img = img.resize((300,300))
img
```
